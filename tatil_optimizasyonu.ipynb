```python
def get_transport_cost(transport_results, city):
    transport_data = transport_results.get(city, {'options': [], 'min_costs': {}})
    
    if not transport_data['options']:
        return float('inf') 

    return min(transport_data['min_costs'].values())
```

```python
def filter_hotels(df, user_inputs):
    print("\nğŸ¨ Oteller filtreleniyor (ÅŸehir, otel puanÄ±, konaklama bÃ¼tÃ§esi)...\n")

    filtered_df = df.copy()

    
    filtered_df = filtered_df[filtered_df['Åehir'].isin(user_inputs['city_choices'])]
    print(f"âœ”ï¸ Åehir filtresi uygulandÄ±: {', '.join(user_inputs['city_choices'])}")

    
    filtered_df = filtered_df[filtered_df['Otel PuanÄ±'] >= user_inputs['min_hotel_rating']]
    print(f"âœ”ï¸ Otel puanÄ± filtresi uygulandÄ±: {user_inputs['min_hotel_rating']} ve Ã¼stÃ¼")

    
    total_accommodation_cost = filtered_df['Gecelik Fiyat (â‚º)'] * user_inputs['num_people'] * user_inputs['num_days']
    filtered_df = filtered_df[total_accommodation_cost <= user_inputs['accommodation_budget']]
    print(f"âœ”ï¸ Konaklama bÃ¼tÃ§esi filtresi uygulandÄ±: {user_inputs['accommodation_budget']} â‚º toplam")

    print(f"\nâœ… Filtreleme sonrasÄ± {len(filtered_df)} otel kaldÄ±.")
    return filtered_df

```

```python
def calculate_custom_score(hotel, user_inputs):
    score = 0

  
    score += hotel['Otel PuanÄ±'] * 100

  
    if user_inputs["activities_selected"]:
        selected_activities = [aktiviteler[int(i)-1] for i in user_inputs["activities_selected"]]
        hotel_activities = str(hotel['Aktiviteler']).split(", ")
        matches = len(set(selected_activities).intersection(hotel_activities))
        mismatches = len(selected_activities) - matches
        score += matches * 20 - mismatches * 10

    
    if user_inputs["hotel_features"]:
        selected_features = [otel_ozellikleri[int(i)-1] for i in user_inputs["hotel_features"]]
        hotel_features = str(hotel['Otel Ã–zellikleri']).split(", ")
        matches = len(set(selected_features).intersection(hotel_features))
        mismatches = len(selected_features) - matches
        score += matches * 15 - mismatches * 5

    
    if user_inputs["max_distance"]:
        if hotel['Merkeze UzaklÄ±k (km)'] <= user_inputs["max_distance"]:
            score += 30
        else:
            score -= 20

  
    if user_inputs["family_trip"] == "evet":
        if hotel['Aile Dostu'] == 'Evet':
            score += 25
        else:
            score -= 15

    
    meal_map = {"1": "Sadece kahvaltÄ±", "2": "AÃ§Ä±k bÃ¼fe", "3": "YarÄ±m pansiyon", "4": "Yemek hizmeti gerekli deÄŸil"}
    if user_inputs["meal_preference"] != "4":  # "Hizmet gerekli deÄŸil" ise deÄŸerlendirme yok
        meal_choice = meal_map[user_inputs["meal_preference"]]
        if hotel['Yemek Hizmeti'] == meal_choice:
            score += 20
        else:
            score -= 10

    
    score += random.uniform(-50, 50)

    return score
```

```python
def check_transport_options(df, user_inputs):
    transport_results = {}

    print("\nğŸš— UlaÅŸÄ±m SeÃ§enekleri DeÄŸerlendiriliyor...")

    for city in user_inputs["city_choices"]:
        city_df = df[df['Åehir'] == city]

        if city_df.empty:
            print(f"\nâš ï¸ {city} iÃ§in otel bulunamadÄ±, ulaÅŸÄ±m seÃ§enekleri kontrol edilmeyecek.")
            continue

        print(f"\nğŸ§­ {city} iÃ§in ulaÅŸÄ±m kontrol ediliyor:")

        transport_options = []
        min_costs_dict = {}

        
        selected_transport_types = user_inputs["transport_choices"] if user_inputs["transport_choices"] else ["1", "2", "3", "4"]
        transport_map = {
            "1": ("UÃ§ak Bilet FiyatÄ± (KiÅŸi BaÅŸÄ±)", "UÃ§uÅŸ SÃ¼resi (Saat)", "UÃ§ak"),
            "2": ("OtobÃ¼s FiyatÄ± (KiÅŸi BaÅŸÄ±)", "OtobÃ¼s SÃ¼resi (Saat)", "OtobÃ¼s"),
            "3": ("Tren Ãœcreti (KiÅŸi BaÅŸÄ±)", "Tren SÃ¼resi (Saat)", "Tren"),
            "4": ("AraÃ§la UlaÅŸÄ±m FiyatÄ±", "AraÃ§la SÃ¼re (Saat)", "AraÃ§")
        }

        for transport_key in selected_transport_types:
            if transport_key not in transport_map:
                continue  # GeÃ§ersiz input

            price_column, time_column, transport_name = transport_map[transport_key]

            
            min_price = city_df[price_column].min()
            min_time = city_df[time_column].min()

            budget_check = (min_price * user_inputs["num_people"]) <= user_inputs["transport_budget"]
            time_check = (user_inputs["max_travel_time"] is None) or (min_time <= user_inputs["max_travel_time"])

            if budget_check and time_check:
                total_min_price = min_price * user_inputs["num_people"]
                print(f"âœ… {transport_name}: Uygun (Min Fiyat: â‚º{total_min_price}, Min SÃ¼re: {min_time} saat)")
                transport_options.append(transport_name)
                min_costs_dict[transport_name] = total_min_price
            else:
                reasons = []
                if not budget_check:
                    reasons.append("BÃ¼tÃ§e yetersiz")
                if not time_check:
                    reasons.append("SÃ¼re aÅŸÄ±mÄ±")
                print(f"âŒ {transport_name}: Uygun deÄŸil ({', '.join(reasons)})")

        if not transport_options:
            print(f"âš ï¸ {city} iÃ§in uygun ulaÅŸÄ±m seÃ§eneÄŸi bulunamadÄ±. LÃ¼tfen bÃ¼tÃ§enizi veya ulaÅŸÄ±m tercihinizi gÃ¶zden geÃ§irin.")

        transport_results[city] = {
            'options': transport_options,
            'min_costs': min_costs_dict
        }

    return transport_results
```

```python
def a_star_algorithm(filtered_df, user_inputs, transport_results):
    print("\nâ­ A* AlgoritmasÄ± Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor...")

    results = {}

    for city in user_inputs["city_choices"]:
        city_hotels = filtered_df[filtered_df['Åehir'] == city]

        if city_hotels.empty:
            print(f"âŒ {city} iÃ§in uygun otel bulunamadÄ±.")
            results[city] = None
            continue

        
        transport_data = transport_results.get(city, {'options': [], 'min_costs': {}})
        if not transport_data['options']:
            print(f"âš ï¸ {city} iÃ§in uygun ulaÅŸÄ±m seÃ§eneÄŸi yok, algoritmadan Ã§Ä±karÄ±lÄ±yor.")
            results[city] = None
            continue

        city_hotels = city_hotels.copy()

        
        min_transport_cost = min(transport_data['min_costs'].values())
        city_hotels['G'] = city_hotels['Gecelik Fiyat (â‚º)'] * user_inputs["num_people"] * user_inputs["num_days"] + min_transport_cost

       
        city_hotels['H'] = city_hotels.apply(lambda hotel: -calculate_custom_score(hotel, user_inputs), axis=1)

        
        city_hotels['Random'] = [random.uniform(-50, 50) for _ in range(len(city_hotels))]

        
        city_hotels['F'] = city_hotels['G'] + city_hotels['H'] - city_hotels['Random']

        
        best_hotel = city_hotels.sort_values(by='F').iloc[0]

        print(f"âœ… {city} iÃ§in en uygun otel bulundu: {best_hotel['Otel AdÄ±']} (â‚º{best_hotel['Gecelik Fiyat (â‚º)']}, â­ {best_hotel['Otel PuanÄ±']}, {best_hotel['Merkeze UzaklÄ±k (km)']} km)")
        results[city] = best_hotel

    return results

```

```python
def genetic_algorithm(filtered_df, user_inputs, transport_results, population_size=20, generations=50, mutation_rate=0.2):
    print("\nğŸ§¬ Genetik Algoritma Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor...")

    results = {}

    for city in user_inputs["city_choices"]:
        city_hotels = filtered_df[filtered_df['Åehir'] == city]

        if city_hotels.empty:
            print(f"âŒ {city} iÃ§in uygun otel bulunamadÄ±.")
            results[city] = None
            continue

        
        transport_data = transport_results.get(city, {'options': [], 'min_costs': {}})
        if not transport_data['options']:
            print(f"âš ï¸ {city} iÃ§in uygun ulaÅŸÄ±m seÃ§eneÄŸi yok, algoritmadan Ã§Ä±karÄ±lÄ±yor.")
            results[city] = None
            continue

        min_transport_cost = min(transport_data['min_costs'].values())

        city_hotels = city_hotels.copy()

       
        population = city_hotels.sample(min(population_size, city_hotels.shape[0]), replace=True)

        for generation in range(generations):
            # Fitness hesapla: Skor - Toplam Maliyet
            population['Fitness'] = population.apply(
                lambda hotel: calculate_custom_score(hotel, user_inputs) - (
                    hotel['Gecelik Fiyat (â‚º)'] * user_inputs["num_people"] * user_inputs["num_days"] + min_transport_cost
                ),
                axis=1
            )

            
            population = population.sort_values(by='Fitness', ascending=False)
            parents = population.head(population_size // 2)

            
            children = []
            while len(children) < population_size - len(parents):
                parent1 = parents.sample(1).iloc[0]
                parent2 = parents.sample(1).iloc[0]

                child = parent1.copy()

                
                if random.random() > 0.5:
                    child['Gecelik Fiyat (â‚º)'] = parent2['Gecelik Fiyat (â‚º)']
                if random.random() > 0.5:
                    child['Merkeze UzaklÄ±k (km)'] = parent2['Merkeze UzaklÄ±k (km)']
                if random.random() > 0.5:
                    child['Otel PuanÄ±'] = parent2['Otel PuanÄ±']

                children.append(child)

            
            for child in children:
                if random.random() < mutation_rate:
                    child['Gecelik Fiyat (â‚º)'] *= random.uniform(0.9, 1.1)
                    child['Merkeze UzaklÄ±k (km)'] *= random.uniform(0.9, 1.1)
                    child['Otel PuanÄ±'] = min(5.0, child['Otel PuanÄ±'] * random.uniform(0.9, 1.1))

            
            population = pd.concat([parents] + children, ignore_index=True)

        
        best_hotel = population.sort_values(by='Fitness', ascending=False).iloc[0]

        print(f"âœ… {city} iÃ§in en uygun otel bulundu: {best_hotel['Otel AdÄ±']} (â‚º{best_hotel['Gecelik Fiyat (â‚º)']}, â­ {best_hotel['Otel PuanÄ±']}, {best_hotel['Merkeze UzaklÄ±k (km)']} km)")
        results[city] = best_hotel

    return results
```

```python
def weighted_csp(filtered_df, user_inputs, transport_results):
    print("\nâš–ï¸ Weighted CSP AlgoritmasÄ± Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor...")

    results = {}

    
    weights = user_inputs["weights"]

    for city in user_inputs["city_choices"]:
        city_hotels = filtered_df[filtered_df['Åehir'] == city]

        if city_hotels.empty:
            print(f"âŒ {city} iÃ§in uygun otel bulunamadÄ±.")
            results[city] = None
            continue

       
        transport_data = transport_results.get(city, {'options': [], 'min_costs': {}})
        if not transport_data['options']:
            print(f"âš ï¸ {city} iÃ§in uygun ulaÅŸÄ±m seÃ§eneÄŸi yok, algoritmadan Ã§Ä±karÄ±lÄ±yor.")
            results[city] = None
            continue

        min_transport_cost = min(transport_data['min_costs'].values())

        city_hotels = city_hotels.copy()

        
        city_hotels['Total Cost'] = (
            city_hotels['Gecelik Fiyat (â‚º)'] * user_inputs['num_people'] * user_inputs['num_days']
            + min_transport_cost
        )

        # Normalizasyon iÃ§in min-max deÄŸerleri
        price_min = city_hotels['Total Cost'].min()
        price_max = city_hotels['Total Cost'].max()
        distance_min = city_hotels['Merkeze UzaklÄ±k (km)'].min()
        distance_max = city_hotels['Merkeze UzaklÄ±k (km)'].max()
        rating_min = city_hotels['Otel PuanÄ±'].min()
        rating_max = city_hotels['Otel PuanÄ±'].max()

        max_activities = city_hotels['Aktiviteler'].apply(lambda x: len(str(x).split(", "))).max()
        max_features = city_hotels['Otel Ã–zellikleri'].apply(lambda x: len(str(x).split(", "))).max()

        def calculate_weighted_score(hotel):
            
            price_score = 1 - ((hotel['Total Cost'] - price_min) / (price_max - price_min + 1e-9))
            
            distance_score = 1 - ((hotel['Merkeze UzaklÄ±k (km)'] - distance_min) / (distance_max - distance_min + 1e-9))
           
            rating_score = (hotel['Otel PuanÄ±'] - rating_min) / (rating_max - rating_min + 1e-9)
            
            activity_count = len(str(hotel['Aktiviteler']).split(", "))
            activities_score = activity_count / (max_activities + 1e-9)
            
            feature_count = len(str(hotel['Otel Ã–zellikleri']).split(", "))
            features_score = feature_count / (max_features + 1e-9)

            
            random_factor = random.uniform(-0.1, 0.1)

            
            total_score = (
                price_score * weights["price"] +
                rating_score * weights["rating"] +
                distance_score * weights["distance"] +
                activities_score * weights["activities"] +
                features_score * weights["features"] +
                random_factor
            )

            return total_score

        
        city_hotels['Weighted_Score'] = city_hotels.apply(calculate_weighted_score, axis=1)

        
        best_hotel = city_hotels.sort_values(by='Weighted_Score', ascending=False).iloc[0]

        print(f"âœ… {city} iÃ§in en uygun otel bulundu: {best_hotel['Otel AdÄ±']} (â‚º{best_hotel['Gecelik Fiyat (â‚º)']}, â­ {best_hotel['Otel PuanÄ±']}, {best_hotel['Merkeze UzaklÄ±k (km)']} km)")
        results[city] = best_hotel

    return results

```

```python
def simulated_annealing(filtered_df, user_inputs, transport_results, initial_temperature=10000, cooling_rate=0.95, iterations_per_temp=10):
    print("\nğŸ”¥ Simulated Annealing AlgoritmasÄ± Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor...")

    results = {}

    for city in user_inputs["city_choices"]:
        city_hotels = filtered_df[filtered_df['Åehir'] == city]

        if city_hotels.empty:
            print(f"âŒ {city} iÃ§in uygun otel bulunamadÄ±.")
            results[city] = None
            continue

        
        transport_data = transport_results.get(city, {'options': [], 'min_costs': {}})
        if not transport_data['options']:
            print(f"âš ï¸ {city} iÃ§in uygun ulaÅŸÄ±m seÃ§eneÄŸi yok, algoritmadan Ã§Ä±karÄ±lÄ±yor.")
            results[city] = None
            continue

        min_transport_cost = min(transport_data['min_costs'].values())

        city_hotels = city_hotels.copy()

        
        temperature = initial_temperature
        current_solution = city_hotels.sample(1).iloc[0]
        best_solution = current_solution

        def fitness(hotel):
            total_cost = hotel['Gecelik Fiyat (â‚º)'] * user_inputs['num_people'] * user_inputs['num_days'] + min_transport_cost
            return calculate_custom_score(hotel, user_inputs) - total_cost

        current_fitness = fitness(current_solution)
        best_fitness = current_fitness

        while temperature > 1:
            for _ in range(iterations_per_temp):
                # KomÅŸu Ã§Ã¶zÃ¼m Ã¼ret (random otel)
                neighbor = city_hotels.sample(1).iloc[0]
                neighbor_fitness = fitness(neighbor)

                # Kabul olasÄ±lÄ±ÄŸÄ± hesapla
                if neighbor_fitness > current_fitness:
                    acceptance_probability = 1
                else:
                    acceptance_probability = math.exp((neighbor_fitness - current_fitness) / (temperature + 1e-9))

                if acceptance_probability > random.random():
                    current_solution = neighbor
                    current_fitness = neighbor_fitness

                    # En iyi Ã§Ã¶zÃ¼m mÃ¼?
                    if current_fitness > best_fitness:
                        best_solution = current_solution
                        best_fitness = current_fitness

            
            temperature *= cooling_rate

        print(f"âœ… {city} iÃ§in en uygun otel bulundu: {best_solution['Otel AdÄ±']} (â‚º{best_solution['Gecelik Fiyat (â‚º)']}, â­ {best_solution['Otel PuanÄ±']}, {best_solution['Merkeze UzaklÄ±k (km)']} km)")
        results[city] = best_solution

    return results

```

```python
import pandas as pd
from IPython.display import display

def display_results(a_star_results, genetic_results, weighted_csp_results, simulated_annealing_results, user_inputs):
    print("\nğŸ“Š Tatil Ã–nerileri KarÅŸÄ±laÅŸtÄ±rmasÄ±\n")

    
    result_rows = []

    for city in user_inputs["city_choices"]:
        print(f"ğŸ“ {city}:")

        def extract_result(algorithm_name, result):
            if result is None:
                print(f"- {algorithm_name}: âŒ Uygun otel bulunamadÄ±.")
                return {
                    "Åehir": city,
                    "Algoritma": algorithm_name,
                    "Otel AdÄ±": "Uygun otel bulunamadÄ±",
                    "Toplam Maliyet (â‚º)": "-",
                    "Otel PuanÄ±": "-",
                    "Merkeze UzaklÄ±k (km)": "-"
                }
            else:
                total_cost = result['Gecelik Fiyat (â‚º)'] * user_inputs['num_people'] * user_inputs['num_days']
                print(f"- {algorithm_name}: {result['Otel AdÄ±']} (Toplam: â‚º{total_cost}, â­ {result['Otel PuanÄ±']}, {result['Merkeze UzaklÄ±k (km)']} km)")
                return {
                    "Åehir": city,
                    "Algoritma": algorithm_name,
                    "Otel AdÄ±": result['Otel AdÄ±'],
                    "Toplam Maliyet (â‚º)": total_cost,
                    "Otel PuanÄ±": result['Otel PuanÄ±'],
                    "Merkeze UzaklÄ±k (km)": result['Merkeze UzaklÄ±k (km)']
                }

        
        result_rows.append(extract_result("A*", a_star_results.get(city)))
        result_rows.append(extract_result("Genetik Algoritma", genetic_results.get(city)))
        result_rows.append(extract_result("Weighted CSP", weighted_csp_results.get(city)))
        result_rows.append(extract_result("Simulated Annealing", simulated_annealing_results.get(city)))

        print("") 

    print("ğŸ‰ Tatil Ã¶nerileri baÅŸarÄ±yla tamamlandÄ±!")

    
    results_df = pd.DataFrame(result_rows)
    results_df.to_csv("tatil_optimizasyon_sonuclari.csv", index=False, encoding='utf-8-sig')

    print("\nğŸ’¾ SonuÃ§lar 'tatil_optimizasyon_sonuclari.csv' dosyasÄ±na kaydedildi!\n")

    
    display(results_df)
```

```python
def calculate_match_score(hotel, user_inputs):
    total_criteria = 0
    matched_criteria = 0

    
    total_criteria += 1
    if hotel['Otel PuanÄ±'] >= user_inputs['min_hotel_rating']:
        matched_criteria += 1

    
    total_criteria += 1
    if hotel['Merkeze UzaklÄ±k (km)'] <= user_inputs['max_distance']:
        matched_criteria += 1

    
    if user_inputs['family_trip'] == 'evet':
        total_criteria += 1
        if hotel['Aile Dostu'].lower() == 'evet':
            matched_criteria += 1

    
    total_criteria += 1
    meal_map = {1: 'Sadece kahvaltÄ±', 2: 'AÃ§Ä±k bÃ¼fe', 3: 'YarÄ±m pansiyon', 4: 'Yemek hizmeti gerekli deÄŸil'}
    selected_meal = meal_map.get(int(user_inputs['meal_preference']), None)
    if selected_meal and str(hotel['Yemek Hizmeti']) == selected_meal:
        matched_criteria += 1


    if user_inputs['activities_selected']:
        total_criteria += 1
        user_activities = set(user_inputs['activities_selected'])
        hotel_activities = set(str(hotel['Aktiviteler']).split(', '))
        if user_activities.issubset(hotel_activities):
            matched_criteria += 1

    
    if user_inputs['hotel_features']:
        total_criteria += 1
        user_features = set(user_inputs['hotel_features'])
        hotel_features = set(str(hotel['Otel Ã–zellikleri']).split(', '))
        if user_features.issubset(hotel_features):
            matched_criteria += 1

    
    if user_inputs['season']:
        total_criteria += 1
        if str(hotel['Mevsim']) == user_inputs['season']:
            matched_criteria += 1

    
    if user_inputs['hotel_types_selected']:
        total_criteria += 1
        if str(hotel['Otel TÃ¼rÃ¼']) in user_inputs['hotel_types_selected']:
            matched_criteria += 1

    if total_criteria == 0:
        return 0  

    return round((matched_criteria / total_criteria) * 100, 2)  
```

```python
def run_vacation_optimizer():
    while True:
        print("\nğŸš€ Tatil Optimizasyonu BaÅŸlÄ±yor...\n")

        start_time = time.time()

        # 1. KullanÄ±cÄ± girdilerini al
        user_inputs = get_user_inputs()
        if user_inputs is None:
            print("âŒ HatalÄ± giriÅŸ yapÄ±ldÄ±. Program sonlandÄ±rÄ±lÄ±yor.")
            return

        # 2. Veri setini yÃ¼kle
        try:
            df = pd.read_csv("tatil_dataset_36000.csv")
            print(f"\nâœ… Veri seti baÅŸarÄ±yla yÃ¼klendi. Toplam {df.shape[0]} otel kaydÄ± mevcut.")
        except FileNotFoundError:
            print("âŒ Hata: Veri seti bulunamadÄ±. LÃ¼tfen dosya yolunu kontrol edin.")
            return

        # 3. Otelleri filtrele
        filtered_df = filter_hotels(df, user_inputs)
        if filtered_df.empty:
            print("âŒ Kriterlerinize uygun otel bulunamadÄ±. Program sonlandÄ±rÄ±lÄ±yor.")
            return

        # 4. UlaÅŸÄ±m kontrolÃ¼ yap
        transport_results = check_transport_options(filtered_df, user_inputs)

        # 5. AlgoritmalarÄ± Ã§alÄ±ÅŸtÄ±r
        print("\nğŸš€ Algoritmalar Ã§alÄ±ÅŸtÄ±rÄ±lÄ±yor...\n")

        algo_times = {}
        results = {}

        # A*
        start_algo = time.time()
        a_star_results = a_star_algorithm(filtered_df, user_inputs, transport_results)
        algo_times["A*"] = time.time() - start_algo
        results["A*"] = a_star_results

        # Genetic
        start_algo = time.time()
        genetic_results = genetic_algorithm(filtered_df, user_inputs, transport_results)
        algo_times["Genetic"] = time.time() - start_algo
        results["Genetic"] = genetic_results

        # Weighted CSP
        start_algo = time.time()
        weighted_csp_results = weighted_csp(filtered_df, user_inputs, transport_results)
        algo_times["Weighted CSP"] = time.time() - start_algo
        results["Weighted CSP"] = weighted_csp_results

        # Simulated Annealing
        start_algo = time.time()
        simulated_annealing_results = simulated_annealing(filtered_df, user_inputs, transport_results)
        algo_times["Simulated Annealing"] = time.time() - start_algo
        results["Simulated Annealing"] = simulated_annealing_results

        # 6. SonuÃ§larÄ± gÃ¶ster
        display_results(
            a_star_results,
            genetic_results,
            weighted_csp_results,
            simulated_annealing_results,
            user_inputs
        )

        # 7. Performans SÃ¼releri
        print("\nâ±ï¸ Algoritma Ã‡alÄ±ÅŸma SÃ¼releri:")
        for algo, duration in algo_times.items():
            print(f"- {algo}: {duration:.2f} saniye")

        # 8. En iyi algoritma: % eÅŸleÅŸme skoru
        print("\nğŸ† En Ä°yi Algoritma (YÃ¼zdelik EÅŸleÅŸme Skoruna GÃ¶re):")
        algo_match_scores = {}

        for algo, algo_result in results.items():
            total_match_score = 0
            count = 0
            for hotel in algo_result.values():
                if hotel is not None:
                    match_score = calculate_match_score(hotel, user_inputs)
                    total_match_score += match_score
                    count += 1
            if count > 0:
                avg_match_score = total_match_score / count
                algo_match_scores[algo] = avg_match_score
                print(f"- {algo}: Ortalama %{avg_match_score:.2f} eÅŸleÅŸme")
            else:
                algo_match_scores[algo] = 0
                print(f"- {algo}: Uygun otel bulunamadÄ±.")

        best_algo = max(algo_match_scores, key=algo_match_scores.get)
        print(f"\nğŸ¥‡ En iyi sonucu veren algoritma: {best_algo}")

        # 9. CSV Ã§Ä±ktÄ±sÄ±
        output_data = []
        for city in user_inputs["city_choices"]:
            row = {"Åehir": city}
            for algo, algo_result in results.items():
                hotel = algo_result.get(city)
                if hotel is not None:
                    total_cost = hotel['Gecelik Fiyat (â‚º)'] * user_inputs["num_people"] * user_inputs["num_days"]
                    match_score = calculate_match_score(hotel, user_inputs)
                    row[f"{algo} Otel"] = hotel['Otel AdÄ±']
                    row[f"{algo} Toplam Maliyet"] = total_cost
                    row[f"{algo} Puan"] = hotel['Otel PuanÄ±']
                    row[f"{algo} UzaklÄ±k"] = hotel['Merkeze UzaklÄ±k (km)']
                    row[f"{algo} EÅŸleÅŸme (%)"] = match_score
                else:
                    row[f"{algo} Otel"] = "Uygun otel bulunamadÄ±"
                    row[f"{algo} Toplam Maliyet"] = "-"
                    row[f"{algo} Puan"] = "-"
                    row[f"{algo} UzaklÄ±k"] = "-"
                    row[f"{algo} EÅŸleÅŸme (%)"] = "-"
            output_data.append(row)

        output_df = pd.DataFrame(output_data)
        output_df.to_csv("tatil_sonuclari.csv", index=False, encoding='utf-8-sig')
        print("\nğŸ’¾ SonuÃ§lar 'tatil_sonuclari.csv' dosyasÄ±na kaydedildi!")

        total_time = time.time() - start_time
        print(f"\nâ±ï¸ Toplam Ã§alÄ±ÅŸma sÃ¼resi: {total_time:.2f} saniye")

        # 10. SonuÃ§larÄ± tablo olarak gÃ¶ster
        print("\nğŸ“Š SonuÃ§lar Tablosu:")
        from IPython.display import display
        display(output_df)

        # 11. Yeni plan sor
        new_plan = input("\nYeni bir tatil planÄ± yapmak ister misiniz? (Evet/HayÄ±r): ").strip().lower()
        if new_plan != "evet":
            print("\nğŸ‰ Tatil Optimizasyonu tamamlandÄ±! Ä°yi tatiller dileriz!")
            break
```
