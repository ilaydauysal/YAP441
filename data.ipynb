```python
import heapq

def calculate_a_star_score(hotel, user_inputs):
    people = user_inputs['kisi_sayisi']
    nights = user_inputs['gun_sayisi']

    # Accommodation cost
    accommodation_cost = hotel['Gecelik Fiyat (₺)'] * nights * people

    # Transport calculation
    transport_cost = None
    transport_time = None

    if 'Uçak' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2
        transport_time = hotel['Uçuş Süresi (Saat)']
    elif 'Tren' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Tren Ücreti (Kişi Başı)'] * people
        transport_time = hotel['Tren Süresi (Saat)']
    elif 'Otobüs' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Otobüs Fiyatı (Kişi Başı)'] * people
        transport_time = hotel['Otobüs Süresi (Saat)']
    elif 'Araç' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Araçla Ulaşım Fiyatı']
        transport_time = hotel['Araçla Süre (Saat)']
    else:  # If user selected "Fark etmez"
        transport_options = [
            (hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2, hotel['Uçuş Süresi (Saat)']),
            (hotel['Tren Ücreti (Kişi Başı)'] * people, hotel['Tren Süresi (Saat)']),
            (hotel['Otobüs Fiyatı (Kişi Başı)'] * people, hotel['Otobüs Süresi (Saat)']),
            (hotel['Araçla Ulaşım Fiyatı'], hotel['Araçla Süre (Saat)'])
        ]
        transport_cost, transport_time = min(transport_options, key=lambda x: x[0])

    # Filter by budget and time
    if accommodation_cost > user_inputs['otel_butcesi'] or transport_cost > user_inputs['ulasim_butcesi']:
        return None

    if user_inputs['max_ulasim_suresi'] is not None and transport_time > user_inputs['max_ulasim_suresi']:
        return None

    # Calculate score
    g_n = accommodation_cost + transport_cost

    # Heuristic: common features + activities + hotel rating
    shared_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_inputs['istenen_ozellikler'])
    shared_activities = sum(activity in hotel['Aktiviteler'] for activity in user_inputs['istenen_aktiviteler'])
    rating_score = hotel['Otel Puanı']

    h_n = -((shared_features + shared_activities) * 100) + ((5 - rating_score) * 50)

    total_score = g_n + h_n

    return total_score


def find_best_hotel_a_star(filtered_hotels, user_inputs):
    heap = []

    for _, hotel in filtered_hotels.iterrows():
        score = calculate_a_star_score(hotel, user_inputs)
        if score is not None:
            heapq.heappush(heap, (score, hotel))

    if not heap:
        return None
    return heapq.heappop(heap)[1]


def run_a_star_algorithm(df, user_inputs):
    results_found = False
    selected_cities = user_inputs['sehir_tercihi'] if user_inputs['sehir_tercihi'] else df['Şehir'].unique().tolist()

    for city in selected_cities:
        print(f"\n🔍 Searching best hotel in {city} (A* Algorithm)...")

        filtered_df = df[df['Şehir'] == city]

        if user_inputs['mevsim'] and 'Tüm Yıl' not in user_inputs['mevsim']:
            filtered_df = filtered_df[filtered_df['Mevsim'].isin(user_inputs['mevsim'] + ['Tüm Yıl'])]


        if user_inputs['aile_tercihi'] == 'Evet':
            filtered_df = filtered_df[filtered_df['Aile Dostu'] == 'Evet']

        if user_inputs['yemek_tercihi'] and 'Yemek hizmeti gerekli değil' not in user_inputs['yemek_tercihi']:
            filtered_df = filtered_df[filtered_df['Yemek Hizmeti'].isin(user_inputs['yemek_tercihi'])]

        if user_inputs['max_uzaklik'] is not None:
            filtered_df = filtered_df[filtered_df['Merkeze Uzaklık (km)'] <= user_inputs['max_uzaklik']]

        filtered_df = filtered_df[filtered_df['Otel Puanı'] >= user_inputs['min_puan']]

        best_hotel = find_best_hotel_a_star(filtered_df, user_inputs)

        if best_hotel is not None:
            results_found = True
            print_best_hotel(best_hotel)  # Output fonksiyonunu da birazdan yazacağız
        else:
            print(f"❌ No suitable hotel found in {city}.")

    if not results_found:
        print("\n🚫 No suitable hotel found in any selected city. Please adjust your preferences.")

    return best_hotel if results_found else None
```

```python
import random

def initialize_population(df, population_size):
    return df.sample(population_size)

def evaluate_fitness(hotel, user_inputs):
    people = user_inputs['kisi_sayisi']
    nights = user_inputs['gun_sayisi']
    accommodation_cost = hotel['Gecelik Fiyat (₺)'] * nights * people

    if 'Uçak' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2
    elif 'Tren' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Tren Ücreti (Kişi Başı)'] * people
    elif 'Otobüs' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Otobüs Fiyatı (Kişi Başı)'] * people
    elif 'Araç' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Araçla Ulaşım Fiyatı']
    else:  # "Fark etmez"
        transport_cost = min(
            hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2,
            hotel['Tren Ücreti (Kişi Başı)'] * people,
            hotel['Otobüs Fiyatı (Kişi Başı)'] * people,
            hotel['Araçla Ulaşım Fiyatı']
        )

    total_cost = accommodation_cost + transport_cost

    if total_cost > user_inputs['otel_butcesi'] + user_inputs['ulasim_butcesi']:
        return float('inf')  # Bütçe aşımı varsa cezalandır

    shared_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_inputs['istenen_ozellikler'])
    shared_activities = sum(activity in hotel['Aktiviteler'] for activity in user_inputs['istenen_aktiviteler'])
    rating_score = hotel['Otel Puanı']

    fitness_score = total_cost - ((shared_features + shared_activities) * 500) - (rating_score * 100)

    return fitness_score

def select_parents(population, fitness_scores):
    sorted_indices = sorted(range(len(fitness_scores)), key=lambda i: fitness_scores[i])
    return population.iloc[sorted_indices[:2]]

def crossover_parents(parent1, parent2):
    child = parent1.copy()
    for column in parent1.index:
        if random.random() > 0.5:
            child[column] = parent2[column]
    return child

def mutate_offspring(child, df):
    mutation_chance = 0.2
    if random.random() < mutation_chance:
        random_hotel = df.sample(1).iloc[0]
        mutate_column = random.choice(df.columns.tolist())
        child[mutate_column] = random_hotel[mutate_column]
    return child

def run_genetic_algorithm(df, user_inputs, population_size=50, generations=20):
    selected_cities = user_inputs['sehir_tercihi'] if user_inputs['sehir_tercihi'] else df['Şehir'].unique().tolist()
    best_overall = None
    best_overall_score = float('inf')

    for city in selected_cities:
        print(f"\n🧬 Searching best hotel in {city} (Genetic Algorithm)...")

        filtered_df = df[df['Şehir'] == city]

        # Apply filters
        if user_inputs['mevsim']:
            filtered_df = filtered_df[filtered_df['Mevsim'].isin(user_inputs['mevsim'] + ['Tüm Yıl'])]
        if user_inputs['aile_tercihi'] == 'Evet':
            filtered_df = filtered_df[filtered_df['Aile Dostu'] == 'Evet']
        if user_inputs['yemek_tercihi'] and 'Yemek hizmeti gerekli değil' not in user_inputs['yemek_tercihi']:
            filtered_df = filtered_df[filtered_df['Yemek Hizmeti'].isin(user_inputs['yemek_tercihi'])]
        if user_inputs['max_uzaklik'] is not None:
            filtered_df = filtered_df[filtered_df['Merkeze Uzaklık (km)'] <= user_inputs['max_uzaklik']]
        filtered_df = filtered_df[filtered_df['Otel Puanı'] >= user_inputs['min_puan']]

        if filtered_df.empty:
            print(f"❌ No suitable hotel found in {city}.")
            continue

        population = initialize_population(filtered_df, population_size)

        for _ in range(generations):
            fitness_scores = [evaluate_fitness(hotel, user_inputs) for _, hotel in population.iterrows()]
            parents = select_parents(population, fitness_scores)
            new_population = []

            while len(new_population) < population_size:
                child = crossover_parents(parents.iloc[0], parents.iloc[1])
                child = mutate_offspring(child, filtered_df)
                new_population.append(child)

            population = pd.DataFrame(new_population)

        # Select best hotel from final population
        final_fitness_scores = [evaluate_fitness(hotel, user_inputs) for _, hotel in population.iterrows()]
        best_index = final_fitness_scores.index(min(final_fitness_scores))
        best_hotel = population.iloc[best_index]

        if min(final_fitness_scores) < best_overall_score:
            best_overall_score = min(final_fitness_scores)
            best_overall = best_hotel

        print_best_hotel(best_hotel)

    if best_overall is None:
        print("\n🚫 No suitable hotel found in any selected city for Genetic Algorithm.")

    return best_overall
```

```python
def normalize_weights(weights):
    total = sum(weights.values())
    if total == 0:
        return {key: 0 for key in weights}  # Hatalı ağırlık verilirse sıfırla
    return {key: value / total for key, value in weights.items()}

def calculate_csp_score(hotel, user_inputs, normalized_weights):
    people = user_inputs['kisi_sayisi']
    nights = user_inputs['gun_sayisi']

    # Cost calculation
    accommodation_cost = hotel['Gecelik Fiyat (₺)'] * nights * people

    if 'Uçak' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2
        transport_time = hotel['Uçuş Süresi (Saat)']
    elif 'Tren' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Tren Ücreti (Kişi Başı)'] * people
        transport_time = hotel['Tren Süresi (Saat)']
    elif 'Otobüs' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Otobüs Fiyatı (Kişi Başı)'] * people
        transport_time = hotel['Otobüs Süresi (Saat)']
    elif 'Araç' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Araçla Ulaşım Fiyatı']
        transport_time = hotel['Araçla Süre (Saat)']
    else:  # "Fark etmez"
        transport_options = [
            (hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2, hotel['Uçuş Süresi (Saat)']),
            (hotel['Tren Ücreti (Kişi Başı)'] * people, hotel['Tren Süresi (Saat)']),
            (hotel['Otobüs Fiyatı (Kişi Başı)'] * people, hotel['Otobüs Süresi (Saat)']),
            (hotel['Araçla Ulaşım Fiyatı'], hotel['Araçla Süre (Saat)'])
        ]
        transport_cost, transport_time = min(transport_options, key=lambda x: x[0])

    total_cost = accommodation_cost + transport_cost

    if total_cost > user_inputs['otel_butcesi'] + user_inputs['ulasim_butcesi']:
        return float('inf')

    if user_inputs['max_ulasim_suresi'] is not None and transport_time > user_inputs['max_ulasim_suresi']:
        return float('inf')

    # Criteria calculation
    shared_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_inputs['istenen_ozellikler'])
    shared_activities = sum(activity in hotel['Aktiviteler'] for activity in user_inputs['istenen_aktiviteler'])
    hotel_rating = hotel['Otel Puanı']

    # Weighted sum (lower is better)
    score = (
        normalized_weights['fiyat'] * total_cost -
        normalized_weights['otel_puani'] * hotel_rating * 100 -
        normalized_weights['otel_ozellikleri'] * shared_features * 100 -
        normalized_weights['aktiviteler'] * shared_activities * 100 +
        normalized_weights['ulasim_suresi'] * transport_time * 10
    )

    return score

def run_weighted_csp_algorithm(df, user_inputs, user_weights):
    normalized_weights = normalize_weights(user_weights)
    selected_cities = user_inputs['sehir_tercihi'] if user_inputs['sehir_tercihi'] else df['Şehir'].unique().tolist()

    best_overall_hotel = None
    best_score = float('inf')

    for city in selected_cities:
        print(f"\n⚖️ Searching best hotel in {city} (Weighted CSP)...")

        filtered_df = df[df['Şehir'] == city]

        # Apply filters
        if user_inputs['mevsim']:
            filtered_df = filtered_df[filtered_df['Mevsim'].isin(user_inputs['mevsim'] + ['Tüm Yıl'])]
        if user_inputs['aile_tercihi'] == 'Evet':
            filtered_df = filtered_df[filtered_df['Aile Dostu'] == 'Evet']
        if user_inputs['yemek_tercihi'] and 'Yemek hizmeti gerekli değil' not in user_inputs['yemek_tercihi']:
            filtered_df = filtered_df[filtered_df['Yemek Hizmeti'].isin(user_inputs['yemek_tercihi'])]
        if user_inputs['max_uzaklik'] is not None:
            filtered_df = filtered_df[filtered_df['Merkeze Uzaklık (km)'] <= user_inputs['max_uzaklik']]
        filtered_df = filtered_df[filtered_df['Otel Puanı'] >= user_inputs['min_puan']]

        if filtered_df.empty:
            print(f"❌ No suitable hotel found in {city}.")
            continue

        # Evaluate all hotels
        filtered_df['score'] = filtered_df.apply(lambda hotel: calculate_csp_score(hotel, user_inputs, normalized_weights), axis=1)

        # Select the best one
        best_hotel = filtered_df.sort_values('score').iloc[0]

        if best_hotel['score'] < best_score:
            best_score = best_hotel['score']
            best_overall_hotel = best_hotel

        print_best_hotel(best_hotel)

    if best_overall_hotel is None:
        print("\n🚫 No suitable hotel found in any selected city for Weighted CSP.")

    return best_overall_hotel
```

```python
import numpy as np
import random

def calculate_sa_score(hotel, user_inputs):
    people = user_inputs['kisi_sayisi']
    nights = user_inputs['gun_sayisi']

    accommodation_cost = hotel['Gecelik Fiyat (₺)'] * nights * people

    if 'Uçak' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2
        transport_time = hotel['Uçuş Süresi (Saat)']
    elif 'Tren' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Tren Ücreti (Kişi Başı)'] * people
        transport_time = hotel['Tren Süresi (Saat)']
    elif 'Otobüs' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Otobüs Fiyatı (Kişi Başı)'] * people
        transport_time = hotel['Otobüs Süresi (Saat)']
    elif 'Araç' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Araçla Ulaşım Fiyatı']
        transport_time = hotel['Araçla Süre (Saat)']
    else:
        transport_options = [
            (hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2, hotel['Uçuş Süresi (Saat)']),
            (hotel['Tren Ücreti (Kişi Başı)'] * people, hotel['Tren Süresi (Saat)']),
            (hotel['Otobüs Fiyatı (Kişi Başı)'] * people, hotel['Otobüs Süresi (Saat)']),
            (hotel['Araçla Ulaşım Fiyatı'], hotel['Araçla Süre (Saat)'])
        ]
        transport_cost, transport_time = min(transport_options, key=lambda x: x[0])

    total_cost = accommodation_cost + transport_cost

    if total_cost > user_inputs['otel_butcesi'] + user_inputs['ulasim_butcesi']:
        return float('inf')

    if user_inputs['max_ulasim_suresi'] is not None and transport_time > user_inputs['max_ulasim_suresi']:
        return float('inf')

    shared_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_inputs['istenen_ozellikler'])
    shared_activities = sum(activity in hotel['Aktiviteler'] for activity in user_inputs['istenen_aktiviteler'])
    hotel_rating = hotel['Otel Puanı']

    score = (
        total_cost
        - (shared_features * 100)
        - (shared_activities * 100)
        - (hotel_rating * 100)
        + (transport_time * 10)
    )

    return score

def generate_neighbor_solution(hotel_list, current_index):
    if len(hotel_list) <= 1:
        return current_index  # Tek otel varsa kendisini döndür
    indices = list(range(len(hotel_list)))
    indices.remove(current_index)
    return random.choice(indices)

def run_simulated_annealing_algorithm(df, user_inputs, initial_temperature=1000, cooling_rate=0.003):
    selected_cities = user_inputs['sehir_tercihi'] if user_inputs['sehir_tercihi'] else df['Şehir'].unique().tolist()

    best_overall_hotel = None
    best_overall_score = float('inf')

    for city in selected_cities:
        print(f"\n🔥 Searching best hotel in {city} (Simulated Annealing)...")

        filtered_df = df[df['Şehir'] == city]

        # Apply filters
        if user_inputs['mevsim']:
            filtered_df = filtered_df[filtered_df['Mevsim'].isin(user_inputs['mevsim'] + ['Tüm Yıl'])]
        if user_inputs['aile_tercihi'] == 'Evet':
            filtered_df = filtered_df[filtered_df['Aile Dostu'] == 'Evet']
        if user_inputs['yemek_tercihi'] and 'Yemek hizmeti gerekli değil' not in user_inputs['yemek_tercihi']:
            filtered_df = filtered_df[filtered_df['Yemek Hizmeti'].isin(user_inputs['yemek_tercihi'])]
        if user_inputs['max_uzaklik'] is not None:
            filtered_df = filtered_df[filtered_df['Merkeze Uzaklık (km)'] <= user_inputs['max_uzaklik']]
        filtered_df = filtered_df[filtered_df['Otel Puanı'] >= user_inputs['min_puan']]

        if filtered_df.empty:
            print(f"❌ No suitable hotel found in {city}.")
            continue

        hotels = filtered_df.reset_index(drop=True)
        current_index = random.randint(0, len(hotels) - 1)
        current_hotel = hotels.iloc[current_index]
        current_score = calculate_sa_score(current_hotel, user_inputs)

        best_city_hotel = current_hotel
        best_city_score = current_score

        temperature = initial_temperature

        while temperature > 1:
            neighbor_index = generate_neighbor_solution(hotels, current_index)
            neighbor_hotel = hotels.iloc[neighbor_index]
            neighbor_score = calculate_sa_score(neighbor_hotel, user_inputs)

            delta_score = neighbor_score - current_score

            if delta_score < 0 or np.exp(-delta_score / temperature) > random.random():
                current_index = neighbor_index
                current_score = neighbor_score
                current_hotel = neighbor_hotel

                if current_score < best_city_score:
                    best_city_score = current_score
                    best_city_hotel = current_hotel

            temperature *= (1 - cooling_rate)

        if best_city_score < best_overall_score:
            best_overall_score = best_city_score
            best_overall_hotel = best_city_hotel

        print_best_hotel(best_city_hotel)

    if best_overall_hotel is None:
        print("\n🚫 No suitable hotel found in any selected city for Simulated Annealing.")

    return best_overall_hotel
```

```python
import numpy as np
import random

def calculate_sa_score(hotel, user_inputs):
    people = user_inputs['kisi_sayisi']
    nights = user_inputs['gun_sayisi']

    accommodation_cost = hotel['Gecelik Fiyat (₺)'] * nights * people

    if 'Uçak' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2
        transport_time = hotel['Uçuş Süresi (Saat)']
    elif 'Tren' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Tren Ücreti (Kişi Başı)'] * people
        transport_time = hotel['Tren Süresi (Saat)']
    elif 'Otobüs' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Otobüs Fiyatı (Kişi Başı)'] * people
        transport_time = hotel['Otobüs Süresi (Saat)']
    elif 'Araç' in user_inputs['ulasim_tercihi']:
        transport_cost = hotel['Araçla Ulaşım Fiyatı']
        transport_time = hotel['Araçla Süre (Saat)']
    else:
        transport_options = [
            (hotel['Uçak Bilet Fiyatı (Kişi Başı)'] * people * 2, hotel['Uçuş Süresi (Saat)']),
            (hotel['Tren Ücreti (Kişi Başı)'] * people, hotel['Tren Süresi (Saat)']),
            (hotel['Otobüs Fiyatı (Kişi Başı)'] * people, hotel['Otobüs Süresi (Saat)']),
            (hotel['Araçla Ulaşım Fiyatı'], hotel['Araçla Süre (Saat)'])
        ]
        transport_cost, transport_time = min(transport_options, key=lambda x: x[0])

    total_cost = accommodation_cost + transport_cost

    if total_cost > user_inputs['otel_butcesi'] + user_inputs['ulasim_butcesi']:
        return float('inf')

    if user_inputs['max_ulasim_suresi'] is not None and transport_time > user_inputs['max_ulasim_suresi']:
        return float('inf')

    shared_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_inputs['istenen_ozellikler'])
    shared_activities = sum(activity in hotel['Aktiviteler'] for activity in user_inputs['istenen_aktiviteler'])
    hotel_rating = hotel['Otel Puanı']

    score = (
        total_cost
        - (shared_features * 100)
        - (shared_activities * 100)
        - (hotel_rating * 100)
        + (transport_time * 10)
    )

    return score

def generate_neighbor_solution(hotel_list, current_index):
    if len(hotel_list) <= 1:
        return current_index  # Tek otel varsa kendisini döndür
    indices = list(range(len(hotel_list)))
    indices.remove(current_index)
    return random.choice(indices)

def run_simulated_annealing_algorithm(df, user_inputs, initial_temperature=1000, cooling_rate=0.003):
    selected_cities = user_inputs['sehir_tercihi'] if user_inputs['sehir_tercihi'] else df['Şehir'].unique().tolist()

    best_overall_hotel = None
    best_overall_score = float('inf')

    for city in selected_cities:
        print(f"\n🔥 Searching best hotel in {city} (Simulated Annealing)...")

        filtered_df = df[df['Şehir'] == city]

        # Apply filters
        if user_inputs['mevsim']:
            filtered_df = filtered_df[filtered_df['Mevsim'].isin(user_inputs['mevsim'] + ['Tüm Yıl'])]
        if user_inputs['aile_tercihi'] == 'Evet':
            filtered_df = filtered_df[filtered_df['Aile Dostu'] == 'Evet']
        if user_inputs['yemek_tercihi'] and 'Yemek hizmeti gerekli değil' not in user_inputs['yemek_tercihi']:
            filtered_df = filtered_df[filtered_df['Yemek Hizmeti'].isin(user_inputs['yemek_tercihi'])]
        if user_inputs['max_uzaklik'] is not None:
            filtered_df = filtered_df[filtered_df['Merkeze Uzaklık (km)'] <= user_inputs['max_uzaklik']]
        filtered_df = filtered_df[filtered_df['Otel Puanı'] >= user_inputs['min_puan']]

        if filtered_df.empty:
            print(f"❌ No suitable hotel found in {city}.")
            continue

        hotels = filtered_df.reset_index(drop=True)
        current_index = random.randint(0, len(hotels) - 1)
        current_hotel = hotels.iloc[current_index]
        current_score = calculate_sa_score(current_hotel, user_inputs)

        best_city_hotel = current_hotel
        best_city_score = current_score

        temperature = initial_temperature

        while temperature > 1:
            neighbor_index = generate_neighbor_solution(hotels, current_index)
            neighbor_hotel = hotels.iloc[neighbor_index]
            neighbor_score = calculate_sa_score(neighbor_hotel, user_inputs)

            delta_score = neighbor_score - current_score

            if delta_score < 0 or np.exp(-delta_score / temperature) > random.random():
                current_index = neighbor_index
                current_score = neighbor_score
                current_hotel = neighbor_hotel

                if current_score < best_city_score:
                    best_city_score = current_score
                    best_city_hotel = current_hotel

            temperature *= (1 - cooling_rate)

        if best_city_score < best_overall_score:
            best_overall_score = best_city_score
            best_overall_hotel = best_city_hotel

        print_best_hotel(best_city_hotel)

    if best_overall_hotel is None:
        print("\n🚫 No suitable hotel found in any selected city for Simulated Annealing.")

    return best_overall_hotel
```

```python
def run_all_algorithms(user_inputs, weights):
    results = []
    selected_cities = user_inputs['sehir_tercihi']

    for city in selected_cities:
        print(f"\n🌍 Processing city: {city}")

        # Şehre göre filtreleme
        city_df = df[df['Şehir'] == city].copy()

        # A* Search
        best_a_star, time_a_star, mem_a_star = measure_performance(lambda: run_a_star_algorithm(city_df.copy(), user_inputs))
        score_a_star = calculate_score(best_a_star, user_inputs)
        results.append(("A* Search", best_a_star, time_a_star, mem_a_star, score_a_star, city))

        # Genetic Algorithm
        best_genetic, time_genetic, mem_genetic = measure_performance(lambda: run_genetic_algorithm(city_df.copy(), user_inputs))
        score_genetic = calculate_score(best_genetic, user_inputs)
        results.append(("Genetic Algorithm", best_genetic, time_genetic, mem_genetic, score_genetic, city))

        # Weighted CSP
        best_csp, time_csp, mem_csp = measure_performance(lambda: run_weighted_csp_algorithm(city_df.copy(), user_inputs, weights))
        score_csp = calculate_score(best_csp, user_inputs)
        results.append(("Weighted CSP", best_csp, time_csp, mem_csp, score_csp, city))

        # Simulated Annealing
        best_sa, time_sa, mem_sa = measure_performance(lambda: run_simulated_annealing_algorithm(city_df.copy(), user_inputs))
        score_sa = calculate_score(best_sa, user_inputs)
        results.append(("Simulated Annealing", best_sa, time_sa, mem_sa, score_sa, city))

    return results
```

```python
import time
import tracemalloc

def measure_performance(func):
    tracemalloc.start()
    start_time = time.time()

    result = func()

    end_time = time.time()
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    exec_time = round(end_time - start_time, 4)
    memory_usage = round(peak / 1024, 2)

    return result, exec_time, memory_usage
```

```python
def calculate_score(hotel, user_inputs):
    if hotel is None:
        return 0

    score = 0
    score += hotel['Otel Puanı'] * 20

    # Güvenlik: Eğer kullanıcı seçim yapmadıysa boş liste ver
    user_features = user_inputs.get('istenen_ozellikler', [])
    user_activities = user_inputs.get('istenen_aktiviteler', [])

    # Özellik eşleşmeleri
    matched_features = sum(feature in hotel['Otel Özellikleri'] for feature in user_features)
    score += matched_features * 2

    # Aktivite eşleşmeleri
    matched_activities = sum(activity in hotel['Aktiviteler'] for activity in user_activities)
    score += matched_activities * 2

    return round(score, 1)

```
